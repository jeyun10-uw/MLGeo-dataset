```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
List = [1,2,3,4,5]
Array = np.array(List)
ArraySquared= Array**2

plt.plot(List,Array,label='Linear')
plt.plot(List,ArraySquared,label='Squared')

plt.title('Figure',fontsize=25)
plt.xlabel('X',fontsize=20)
plt.ylabel('Y',fontsize=20)
plt.xticks([1,2,3,4,5],fontsize=15)
plt.yticks([],fontsize=15)
plt.legend(prop={'size':15})
```
